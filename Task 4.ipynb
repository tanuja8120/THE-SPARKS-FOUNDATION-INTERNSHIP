```python
df.rename(columns={'iyear':'Year','imonth':'Month','extended':'Extended','iday':'Day','country_txt':'Country',
                   'provstate':'State','region_txt':'Region','attacktype1_txt':'AttackType','target1':'Target',
                   'nkill':'Killed','nwound':'Wounded','summary':'Summary','gname':'Group','targtype1_txt':'Target_type',
                   'weaptype1_txt':'Weapon_type','motive':'Motive','city':'City','latitude':'Latitude','longitude':'Longitude'},inplace=True)
df.head()
```

```python
df=df[ ['Year','Month','Day','Country','State','Region','AttackType','Target','Killed','Wounded','Group','Target_type','Weapon_type','Motive','City','Latitude','Longitude','Summary']]
```

```python
#identifying number of nullvalues
df.isnull().sum()
```

```python
# Identifying Countries that have been most attacked 
df.Country.value_counts().head(15)
```

```python
#visualising the top ten counties attacked
plt.figure(figsize=(10,5))
plt.title('top ten countries attacked')
sns.barplot(df.Country.value_counts().head(10).index,df.Country.value_counts().head(10))
plt.xlabel('countries attacked')
plt.xticks(rotation=90)
plt.show()
```

```python
df.AttackType.value_counts().head(10)
```

```python
#visualising  the most attacktypes took place 
plt.figure(figsize=(10,5))
plt.title('top ten attack mode')
sns.barplot(df.AttackType.value_counts().head(10).index,df.AttackType.value_counts().head(10))
plt.xlabel('attacktype')
plt.xticks(rotation=90)
plt.show()
```

```python
df.Year.value_counts().head(10)
```

```python
#visualising the years in which  most o terrorist attack tok place
plt.figure(figsize=(10,5))
plt.title('years with most terrorist attacks')
sns.barplot(df.Year.value_counts().head(10).index,df.Year.value_counts().head(10))
plt.xlabel('years')
plt.xticks(rotation=90)
plt.show()
```

```python
df.Weapon_type.value_counts().head(10)
```

```python
#visualising the most of the weapon type used
plt.figure(figsize=(10,5))
plt.title('Weapon_type used')
sns.barplot(df.Weapon_type.value_counts().head(10).index,df.Weapon_type.value_counts().head(10))
plt.xlabel('weapon_type')
plt.xticks(rotation=90)
plt.show()
```

```python
sns.pairplot(df,hue='Region',palette='husl',height=3)
plt.show()
```
