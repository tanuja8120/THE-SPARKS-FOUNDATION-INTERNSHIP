```python
#visualising the data

plt.figure(figsize=(10,5))
plt.title('Matches conducted in cities')
sns.countplot(x='city',data=df)
plt.xlabel('City')
plt.xticks(rotation=90)
plt.show()
```

```python
# Most of the matches conducted in mumbai city
```

```python
df.winner.value_counts()
```

```python
plt.figure(figsize=(10,5))
plt.title('Winners of the match ')
sns.countplot(x='winner',data=df)
plt.xlabel('winner')
plt.xticks(rotation=90)
plt.show()
```

```python
df.win_by_runs.value_counts()
```

```python
max_winner=df.groupby(['winner'],as_index=False)['win_by_runs'].max()
plt.figure(figsize=(10,5))
plt.xticks(rotation=90)
sns.barplot(max_winner['winner'],max_winner['win_by_runs'])
```

### mumbaiindians are the winners most of the time

```python
plt.figure(figsize = (10,5))
sns.barplot(x= df['player_of_match'].value_counts()[:5].index, y= df['player_of_match'].value_counts()[:5])
plt.title('Player Of The match')
plt.xlabel('Player')
plt.ylabel('Matches')
plt.xticks(rotation = 90)
```

```python
plt.figure(figsize=(10,5))
plt.bar(list(df['toss_winner'].value_counts().keys()),list(df['toss_winner'].value_counts()))
plt.xticks(rotation = 90)
plt.show()
```

```python
plt.subplots(figsize=(15,5))
sns.countplot(x ='venue',data = df,order =df['venue'].value_counts().index)
plt.title('Venues')
plt.xlabel('Stadiums')
plt.ylabel('Matches')
plt.xticks(rotation = 90)
```
